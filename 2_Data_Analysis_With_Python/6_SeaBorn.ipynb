## Data Visualization With Seaborn

Seaborn is a python visualization library based on Matplotlib that provides a high-level interface for drawing attractive and informative statistical graphics . Seaborn helps in creating complex visualization with just a few lines of code. In this notebook we will cover the basics of seaborn , including creating various types of plots and customizing them.

```python
%matplotlib inline
import seaborn as sns
```

```python
## Basic plotting with seaborn
tips = sns.load_dataset('tips')
tips
```

```python
## creating a scatter plot

import matplotlib.pyplot as plt

sns.scatterplot(x='total_bill',y='tip',data=tips)
plt.title("Scatter plot of total Bill vs Tip")
plt.show()
```

```python
## Line Plot

sns.lineplot(x='size',y='total_bill',data=tips)
plt.title("Line Plot of total bill vs size")
plt.show()
```

```python
## Categorical Plots

## Bar plot

sns.barplot(x='day',y='total_bill',data=tips)
plt.title('Bar Plot of Total Bill by Day')
plt.show()
```

```python
## Box Plot
sns.boxplot(x='day',y='total_bill',data=tips)
plt.title("Box Plot of total bill by day")
plt.show()
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
plt.title("voilin Plot of total bill by day")
plt.show()
```

```python
### Histogram

sns.histplot(tips['total_bill'],bins=10,kde=True)
```

```python
### KDE Plot

sns.kdeplot(tips['total_bill'],fill=True)
```

```python
### Pair plot

sns.pairplot(tips)
```

```python
### Heatmap

corr = tips[['total_bill','tip','size']].corr()
corr
```

```python
sns.heatmap(corr,annot=True,cmap='coolwarm')
```

```python
import pandas as pd
df = pd.read_csv("sales_data.csv")
df.head(5)
```

```python
## Plot total sales by product

plt.figure(figsize=(10,6))
sns.barplot(x='Product Category',y="Total Revenue",data = df,estimator=sum)
plt.title('Total Sales by Product')
plt.xlabel('product')
plt.ylabel('tatal sales')
plt.show()
```

```python
# Plot total sales by Region

plt.figure(figsize=(10,6))
sns.barplot(x='Region',y="Total Revenue",data = df,estimator=sum)
plt.title('Total Sales by Region')
plt.xlabel('Region')
plt.ylabel('tatal sales')
plt.show()
```
